## Count Frequency in a range




```python
from typing import *

def countFrequency(n: int, m: int, edges: List[List[int]]):
    mp = {}
    hash_og = {}
    for i in edges:
        if i not in hash_og:
            hash_og[i] = 1
        else:
            hash_og[i] += 1
    for i in range(1, n+1):
        if i not in hash_og:
            mp[i] = 0
        else:
            mp[i] = hash_og[i]
    return  mp.values()
```

## Highest / Lower Frequencies




```python
from typing import List

def getFrequencies(v: List[int]) -> List[int]: 
    # Write your code here
    mp = {}
    for i in v:
        if i not in mp:
            mp[i] = 1
        else:
            mp[i] += 1
    count_hash = []
    for i, val in mp.items():
        count_hash.append((val, i))
    arr = []
    arr1= []
    val_arr = mp.values()
    min_val = min(val_arr)
    max_val = max(val_arr)
    for i, val in count_hash:
        if i == min_val:
            arr.append(val)
        if i == max_val:
            arr1.append(val)
    return [min(arr1), min(arr)]
```
